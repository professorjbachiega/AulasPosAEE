```python
# =====================================================
# Demonstração de Média, Mediana e Moda
# Autor: Exemplo para aula de Estatística
# =====================================================

import numpy as np
import matplotlib.pyplot as plt
from collections import Counter


# Função para calcular a moda
def calcular_moda(dados):
    contagem = Counter(dados)
    maior_frequencia = max(contagem.values())
    modas = [valor for valor, freq in contagem.items()
             if freq == maior_frequencia]
    return modas


# Função para mostrar resultados
def analisar_dados(nome, dados):
    media = np.mean(dados)
    mediana = np.median(dados)
    moda = calcular_moda(dados)

    print("="*50)
    print(nome)
    print("="*50)
    print("Dados:", dados)
    print(f"Média: {media:.2f}")
    print(f"Mediana: {mediana:.2f}")
    print(f"Moda(s): {moda}")
    print()


    # Gráfico
    plt.figure(figsize=(10,4))

    plt.scatter(range(len(dados)), dados,
                s=120, label="Dados")

    plt.axhline(media,
                linestyle="--",
                linewidth=2,
                label=f"Média = {media:.2f}")

    plt.axhline(mediana,
                linestyle="-.",
                linewidth=2,
                label=f"Mediana = {mediana:.2f}")

    for m in moda:
        plt.axhline(m,
                    linestyle=":",
                    linewidth=2,
                    label=f"Moda = {m}")

    plt.title(nome)
    plt.xlabel("Observações")
    plt.ylabel("Valores")
    plt.grid(True)
    plt.legend()
    plt.show()


# =====================================================
# Exemplo 1 - Dados equilibrados
# Média e mediana são parecidas
# =====================================================

dados1 = [5, 6, 7, 8, 9, 10, 10]

analisar_dados(
    "Exemplo 1: Dados equilibrados",
    dados1
)


# =====================================================
# Exemplo 2 - Presença de um valor extremo
# A média é puxada pelo outlier
# =====================================================

dados2 = [5, 6, 7, 8, 9, 10, 100]

analisar_dados(
    "Exemplo 2: Efeito de um valor extremo",
    dados2
)


# =====================================================
# Exemplo 3 - Moda muito evidente
# O valor mais frequente representa a moda
# =====================================================

dados3 = [2, 3, 3, 3, 4, 5, 8, 9]

analisar_dados(
    "Exemplo 3: Identificando a moda",
    dados3
)
```
