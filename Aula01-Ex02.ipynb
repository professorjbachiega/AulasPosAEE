```python
# =====================================================
# Demonstração de Medidas de Dispersão
# Amplitude, Variância, Desvio Padrão e CV
# =====================================================

import numpy as np
import matplotlib.pyplot as plt


# Função para calcular e mostrar as medidas
def analisar_dispersao(nome, dados):

    media = np.mean(dados)
    amplitude = np.max(dados) - np.min(dados)
    variancia = np.var(dados)
    desvio_padrao = np.std(dados)
    cv = (desvio_padrao / media) * 100

    print("="*60)
    print(nome)
    print("="*60)
    print("Dados:", dados)
    print(f"Média: {media:.2f}")
    print(f"Amplitude: {amplitude:.2f}")
    print(f"Variância: {variancia:.2f}")
    print(f"Desvio Padrão: {desvio_padrao:.2f}")
    print(f"Coeficiente de Variação: {cv:.2f}%")
    print()


    # Gráfico da distribuição dos dados
    plt.figure(figsize=(10, 4))

    plt.scatter(range(len(dados)), dados,
                s=120, label="Dados")

    plt.axhline(media,
                linestyle="--",
                linewidth=2,
                color="red",
                label=f"Média = {media:.2f}")

    plt.fill_between(
        range(len(dados)),
        media - desvio_padrao,
        media + desvio_padrao,
        alpha=0.2,
        color="red",
        label="Faixa de 1 Desvio Padrão"
    )

    plt.title(nome)
    plt.xlabel("Observações")
    plt.ylabel("Valores")
    plt.grid(True)
    plt.legend()
    plt.show()


# =====================================================
# Exemplo 1 - Dados pouco dispersos
# Valores próximos da média
# =====================================================

dados1 = [48, 50, 51, 49, 52, 50, 50]

analisar_dispersao(
    "Exemplo 1: Baixa dispersão",
    dados1
)


# =====================================================
# Exemplo 2 - Dados muito dispersos
# Valores distantes da média
# =====================================================

dados2 = [10, 30, 45, 50, 70, 90, 110]

analisar_dispersao(
    "Exemplo 2: Alta dispersão",
    dados2
)


# =====================================================
# Exemplo 3 - Comparação pelo Coeficiente de Variação
# Mesma dispersão relativa, escalas diferentes
# =====================================================

dados3 = [100, 110, 90, 95, 105, 100]

analisar_dispersao(
    "Exemplo 3: Dispersão relativa",
    dados3
)
```
